```python
import pandas as pd
import numpy as np
from prisma import Prisma

db = Prisma()
db.connect()
```

```python
vehicles = db.vehicle.find_many()
```

```python
vehicles = [vehicle.dict() for vehicle in vehicles]
```

```python
df = pd.DataFrame.from_records(vehicles)
```

```python
df = df.filter(items=['year', 'make', 'model', 'price', 'city'])
df = df[df['year'] > 2014]
```

```python
tj = df[df['city'] == 'Tijuana, BC']
tj
```

```python
dd_tj = df[df['price'] <= 30]

for index, row in dd_tj.iterrows():
    #print([index, row['make']])
    #print(row['price'] * 16.35)
    try:
        df.iat[index, 3] = int(row['price'] * 16.35 * 1000)
    except:
        pass
dolar_tj = df[df['price'] <= 20000]
dolar_tj = dolar_tj[dolar_tj['price'] >= 100]


for index, row in dolar_tj.iterrows():
    try:
        df.iat[index, 3] = int(row['price'] * 16.35)
    except:
        pass
    
for index, row in dolar_tj.iterrows():
    model = row['model']
    if model:
        model = model.split(' ')[0]
        #print(model)
        if index < len(df):
            df.iat[index, 2] = model
  
```

```python
df
```

```python
df['ymmc'] = df['year'].astype(str) + ' ' + df['make'] + ' ' + df['model'].str.strip() + ' ' + df['city']
df['ymm'] = df['year'].astype(str) + ' ' + df['make'] + ' ' + df['model'].str.strip() + ' '

df
df
```

```python
ymm = pd.DataFrame(df['ymmc'].unique())
ymm.columns= ['ymmc']
ymm['median'] = np.zeros(len(ymm))
ymm['ymm'] = ''
ymm['city'] = ''

for index, row in ymm.iterrows():
    cs = df[df['ymmc']==row['ymmc']]
    if len(cs) > 0:
        match = cs.iloc[0]['ymm']
        city = cs.iloc[0]['city']
        #print(city)
        ymm.iat[index, 2] = match
        ymm.iat[index, 3] = city
    
ymm
```

```python
for index, row in ymm.iterrows():
    ymm_str = row['ymmc']
    #print(ymm_str)
    
    #print(df[df['ymm'] == ymm_str]['price'].mean())
    #break
    price_col = df[df['ymmc'] == ymm_str]['price']
    
    ymm.iat[index, 1] = price_col.median()
    
    #[price_col.median(), price_col.mean()]



pd.set_option('display.max_rows', len(ymm))
ymm

```
